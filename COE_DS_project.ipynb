# Unemployement Data

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/Unemployment_in_India.csv")
df.dropna(inplace=True)
df
```

```python
X = df['Area']
Y = df[' Unemployment Rate']

plt.bar(X,Y, color = "blue")
plt.title("Area Wise Unemployment")
plt.xlabel("Overall Percentage")

plt.show()
```

**Employment Rate**

```python
Y1 = df[' Estimated_Employed']
X1 = df['Region']

plt.barh(X1,Y1)
plt.title("Employment Rate")
plt.xlabel("Employment Count")
plt.ylabel("States")
plt.show()
```

**Unemployment Rate**

```python
X2 = df['Region']
Y2 = df[' Unemployment Rate']

plt.barh(X2, Y2)
plt.title("Unemployment Rate")
plt.xlabel("Percentage")
plt.ylabel("States")
plt.show()
```

**Labour Rate**

```python
x = df['Region']
y = df[' Estimated_Labour_Participation_Rate']

plt.barh(x,y)
plt.title("Labour Participation Rate")
plt.xlabel("Participation Rate")
plt.ylabel("States")
plt.plot()
```
